```python
import numpy as np
from PIL import Image
import numpy as np
from IPython import display
```

```python
# 変換したい写真の名前
display.Image("./xxxx.JPG")
```

```python
img = np.array(Image.open('./sign.JPG').convert('RGBA'))
```

```python
g=img[:,:,1]>100
g=g.astype(np.int)
b=img[:,:,2]>100
b=b.astype(np.int)
img[:,:,3][g+b==2]=0
```

```python
created_img = Image.fromarray(img)
print(created_img.mode)
created_img.save('./created.png')
```
